```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Creating a barplot
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```

```
# Configuring the style, color palette, font, and font size of the plot
sns.set(style='dark', palette='dark', font="Verdana", font_scale=1)

# Creating a barplot with the new settings
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```

```
# Resetting the settings to the default values
sns.reset_orig()

# Creating a barplot with the default settings
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```
